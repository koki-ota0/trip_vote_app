```python
from flask import Flask, render_template, request
app = Flask(__name__)

# 行き先の選択肢
destinations = [
    {'name': '東京', 'votes': 0},
    {'name': '京都', 'votes': 0},
    {'name': '沖縄', 'votes': 0},
    {'name': '北海道', 'votes': 0}
]

# 投票ページの表示
@app.route('/')
def index():
    return render_template('index.html', destinations=destinations)

# 投票処理
@app.route('/vote', methods=['POST'])
def vote():
    destination_name = request.form['destination']
    for destination in destinations:
        if destination['name'] == destination_name:
            destination['votes'] += 1
    return render_template('results.html', destinations=destinations)

# 結果ページの表示
@app.route('/results')
def results():
    return render_template('results.html', destinations=destinations)

if __name__ == '__main__':
    app.run()
```
